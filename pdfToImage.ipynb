```python
import os
import fitz  # PyMuPDF

# Function to convert PDF to images using PyMuPDF
def convert_pdf_to_images(pdf_path, output_folder):
    os.makedirs(output_folder, exist_ok=True)  # Create the output folder if it doesn't exist
    pdf_document = fitz.open(pdf_path)

    for page_number in range(pdf_document.page_count):
        page = pdf_document[page_number]
        image = page.get_pixmap()
        image.save(os.path.join(output_folder, f"page_{page_number + 1}.png"))

    pdf_document.close()

# Example Usage
convert_pdf_to_images("/content/drive/MyDrive/resource/documents/sample.pdf", "/content/drive/MyDrive/resource/documents/output_images")
```

```python
import os

# Function to convert single-page PDF to an image using PyMuPDF
def convert_pdf_to_image(pdf_path, output_folder):
    os.makedirs(output_folder, exist_ok=True)  # Create the output folder if it doesn't exist
    pdf_document = fitz.open(pdf_path)

    # Since it's a single-page PDF, use page_number = 0
    page_number = 0
    page = pdf_document[page_number]

    # Naming the image based on the PDF file name
    image_name = f"{os.path.splitext(os.path.basename(pdf_path))[0]}.png"
    image_path = os.path.join(output_folder, image_name)

    # Save the image
    image = page.get_pixmap()
    image.save(image_path)

    pdf_document.close()

# Folder paths
pdf_folder = "/content/drive/MyDrive/resource/public_transport/documents"
output_base_folder = "/content/drive/MyDrive/resource/public_transport/images"

# Iterate through each PDF file in the input folder
for pdf_file in os.listdir(pdf_folder):
    if pdf_file.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder, pdf_file)
        convert_pdf_to_image(pdf_path, output_base_folder)
        print(f"Converted {pdf_file} to an image in {output_base_folder}")
```

Resize the image

```python
from PIL import Image

def resize_image(image_path, output_path, target_size=(224, 224)):
    img = Image.open(image_path)
    img = img.resize(target_size, Image.LANCZOS)
    img.save(output_path)

# Example Usage
resize_image("/content/drive/MyDrive/resource/documents/output_images/public_transport/118NP8.png", "/content/drive/MyDrive/resource/documents/output_images/resized_image.png", target_size=(224, 224))
```

Resize in loop

```python
from PIL import Image, ImageOps
import os

def resize_images_in_folder(input_folder, output_folder, target_size=(224, 224)):
    os.makedirs(output_folder, exist_ok=True)

    for image_name in os.listdir(input_folder):
        if image_name.endswith(('.png', '.jpg', '.jpeg')):
            input_path = os.path.join(input_folder, image_name)
            output_path = os.path.join(output_folder, image_name)
            resize_image(input_path, output_path, target_size)

# Example Usage
input_images_folder = "/content/drive/MyDrive/resource/public_transport/images"
output_resized_images_folder = "/content/drive/MyDrive/resource/public_transport/resized_images"

resize_images_in_folder(input_images_folder, output_resized_images_folder, target_size=(224, 224))
```

```python
from PIL import Image, ImageOps
import numpy as np
import os

# Function to normalize pixel values in an image
def normalize_image(image_path, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    img = Image.open(image_path)
    img_array = np.array(img, dtype=np.float32) / 255.0  # Normalize pixel values to [0, 1]

    # Additional normalization steps if needed
    # e.g., img_array = (img_array - mean) / std

    normalized_img = Image.fromarray((img_array * 255).astype(np.uint8))
    normalized_path = os.path.join(output_folder, os.path.basename(image_path))
    normalized_img.save(normalized_path)

# Example Usage
input_resized_images_folder = "/content/drive/MyDrive/resource/public_transport/resized_images"
output_normalized_images_folder = "/content/drive/MyDrive/resource/public_transport/normalized_images"

# Iterate through each resized image in the input folder
for image_name in os.listdir(input_resized_images_folder):
    if image_name.endswith(('.png', '.jpg', '.jpeg')):
        resized_image_path = os.path.join(input_resized_images_folder, image_name)
        normalize_image(resized_image_path, output_normalized_images_folder)
        print(f"Normalized {image_name} and saved in {output_normalized_images_folder}")
```

Gray scale conversion

```python
from PIL import Image
import os

# Function to convert an image to grayscale
def convert_to_grayscale(image_path, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    img = Image.open(image_path)
    grayscale_img = img.convert("L")  # Convert to grayscale
    grayscale_path = os.path.join(output_folder, os.path.basename(image_path))
    grayscale_img.save(grayscale_path)

# Example Usage
input_images_folder = "/content/drive/MyDrive/resource/public_transport/normalized_images"
output_grayscale_images_folder = "/content/drive/MyDrive/resource/public_transport/grayscale_images"

# Iterate through each image in the input folder
for image_name in os.listdir(input_images_folder):
    if image_name.endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(input_images_folder, image_name)
        convert_to_grayscale(image_path, output_grayscale_images_folder)
        print(f"Converted {image_name} to grayscale and saved in {output_grayscale_images_folder}")
```

Data Augmentaion

```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import os
from PIL import Image

# Function to perform data augmentation on an image
def augment_images(input_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    # Iterate through each image in the input folder
    for image_name in os.listdir(input_folder):
        if image_name.endswith(('.png', '.jpg', '.jpeg')):
            image_path = os.path.join(input_folder, image_name)

            # Load the image and convert to a numpy array
            img = Image.open(image_path)
            img_array = np.array(img)

            # Add batch and channel dimensions to make it (1, height, width, channels)
            img_array = np.expand_dims(img_array, axis=0)
            img_array = np.expand_dims(img_array, axis=-1)

            # Generate augmented images and save them to the output folder
            for batch in datagen.flow(img_array, batch_size=1, save_to_dir=output_folder, save_prefix='aug', save_format='png'):
                pass

# Example Usage
input_images_folder = "/content/drive/MyDrive/resource/public_transport/grayscale_images"
output_augmented_images_folder = "/content/drive/MyDrive/resource/public_transport/augmented_images"

augment_images(input_images_folder, output_augmented_images_folder)
```

```python
from PIL import Image, ImageOps
import os

# Function to perform histogram equalization on an image
def histogram_equalization(image_path, output_folder):
    os.makedirs(output_folder, exist_ok=True)

    img = Image.open(image_path)
    img = ImageOps.exif_transpose(img)  # Handle image rotation based on EXIF data
    img = img.convert('L')  # Convert to grayscale
    img = ImageOps.equalize(img)
    equalized_path = os.path.join(output_folder, os.path.basename(image_path))
    img.save(equalized_path)

# Example Usage

input_images_folder = "/content/drive/MyDrive/resource/public_transport/augmented_images"
output_equalized_images_folder = "/content/drive/MyDrive/resource/public_transport/equalized_imagess"

# Iterate through each image in the input folder
for image_name in os.listdir(input_images_folder):
    if image_name.endswith(('.png', '.jpg', '.jpeg')):
        image_path = os.path.join(input_images_folder, image_name)
        histogram_equalization(image_path, output_equalized_images_folder)
        print(f"Equalized histogram for {image_name} and saved in {output_equalized_images_folder}")
```
